```python
from IPython.display import clear_output
import random
```

```python
def player_input():
    
    # checking player input x / o
    
    marker = input('choose X or O? ').lower()
    if marker not in ['o', 'x']:
        clear_output()
        print('Wrong marker!')
        print('Please choose again.')
        player_input()
```

```python
def choose_first():
    
    # who starts first? x or o?
    
    if random.randint(1, 2) == 1:
        t = 'X'
    else:
        t = 'O'
    return t
```

```python
def intoducing_board():
    
    # showing an overall view of board to players
    
    print('This is your board.')
    print('')
    print('       7|8|9')
    print('       -----')
    print('       4|5|6')
    print('       -----')
    print('       1|2|3')
```

```python
def display_board():
    
    # how board is made
    
    board = [
        {'c1': 7, 'c2': 8, 'c3': 9},
        {'c1': 4, 'c2': 5, 'c3': 6},
        {'c1': 1, 'c2': 2, 'c3': 3},
    ]
    return board
```

```python
def player_choice(t , s):
    
    #asking to enter a position
    
    if t == 'X':
        x = 1
    else:
        x = 2
    position = input('Player' + str(x) + ' ,Choose a position? (1 _ 9) ')
    clear_output()
    
    #checking if input position is in (1 ,9)
    # if its not a number or out of range or a place occupied already , get input again
    
    if int(position) in range(1, 10) and int(position) not in s:
        s.append(int(position))
        return int(position)
    else:
        print('Wrong number!')
        player_choice(t , s)
```

```python
def place_marker(board, turn, position):
    
    #putting the turn marker in the correct position in board
    
    for i in range(3):
        for j in range(1,4):
            if board[i]['c' + str(j)] == position :
                #board[position]=marker
                board[i]['c' + str(j)] = turn
                return board
        
```

```python
def changed_board(board):
    # printing the changed board
    print('      ' + str(board[0]['c1']) +' | '+ str(board[0]['c2']) +' | '+ str(board[0]['c3']))
    print('      -----------')
    print('      ' + str(board[1]['c1']) +' | '+ str(board[1]['c2']) +' | '+ str(board[1]['c3']))
    print('      -----------')
    print('      ' + str(board[2]['c1']) +' | '+ str(board[2]['c2']) +' | '+ str(board[2]['c3']))
```

```python
def win_check(board, turn):
    
    # checking lines
    for j in range(3):
        for i in range(1,4):
            if board[j]['c' + str(i)] != turn:
                break
        else:
            print('\n***************')
            print(' player ' + turn + ' won! ')
            print('***************')
            return 'DONE'
        
    # checking columns
    for j in range(1,4):
        for i in range(3):
            if board[i]['c' + str(j)] != turn:
                break
        else:
            print('\n***************')
            print(' player ' + turn + ' won! ')
            print('***************')
            return 'DONE'

    # checking the diameters
    
    # Diameter 1 , 5 , 9
    if board[2]['c1'] == board[1]['c2'] == board[0]['c3'] == turn:
        print('\n***************')
        print(' player ' + turn + ' won! ')
        print('***************')
        return 'DONE'
    
    # Diameter 3 , 5 , 7
    if board[2]['c3'] == board[1]['c2'] == board[0]['c1'] == turn:
        print('\n***************')
        print(' player ' + turn + ' won!')
        print('***************')
        return 'DONE'
```

```python
def change_turn(t):
    
    #exhanging the turn
    #if its X , it changes to O
    #if its O , it changes to X
    
    if t == 'X':
        return 'O'
    else:
        return 'X'
```

```python
def replay():
    
    #asking if players want to play again
    
    q = input('Do you want to play again? yes or no? ').lower()
    if q == 'no':
        clear_output()
        print('Good job!')
        return False
    
    elif q == 'yes':
        clear_output()
        print('Good to play again!')
        
    #if players entered neither yes nor no
    else:
        clear_output()
        print('\nWrong input!')
        replay()
```

```python
print('Welcome to tic tac toe')
print('First of all , one player should choose a marker btw O or X.')
print('Every choice you take, \nthe other marker will be assigned to another player automatically.')

while True:
    player_input()
    clear_output()
    t = choose_first()
    
    if t == 'X':
        print('X will go first')
    else:
        print('O will go first')
        
    intoducing_board()
    b = display_board()
    s = []
    for _ in range(9):
        p = player_choice(t ,s)
        b = place_marker(b, t, p)
        changed_board(b)
        f = win_check(b, t)
        if f == 'DONE':
            break
        t = change_turn(t)
    else:
        print('Both are equal!')
        
    replay()
```
